```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
```

```python
url = r'https://www.hubertiming.com/results/2018mlk'
html = urlopen(url)
```

```python
soup = bs(html)
```

```python
print(soup.title.text)
```

```python
links = soup.find_all('a', href=True)
for link in links :
    print(link['href'])
```

```python

data = []
all_rows = soup.find_all('tr')
first_row = all_rows[5].find_all('td')

l = []
for tag in first_row :
    for x in tag :
        l.append(x.strip())

l
```

```python
soup.find_all('th')
```

```python
headers = []
for x in soup.find_all('th') :
    headers.append(x.text)
```

```python
data = []
for c in range(len(all_rows)) :
    if c < 5 : continue
    row = all_rows[c].find_all('td')
    l = []
    for tag in row :
        for x in tag :
            l.append(x.strip())
    data.append(l)
    
data
```

```python
df = pd.DataFrame(data, columns=headers)
df
```

## From iplt20.com

```python
url_bat = r'https://www.iplt20.com/stats/2021'
html_bat = urlopen(url_bat)
soup_bat = bs(html_bat)
```

```python
soup_bat.title
```

```python
headers_bat = []
for head in soup_bat.find_all('th')[:-7] :
    headers_bat.append(head.text)
headers_bat
```

```python

data_bat = []

for c in range(len(soup_bat.find_all('tr'))) :
    if c==0 or c>=len(soup_bat.find_all('tr'))-7 : continue
    row = soup_bat.find_all('tr')[c].find_all('td')
    l = []
    for x in row :
        l.append(x.text.strip())
    data_bat.append(l)
```

```python
batting_data = pd.DataFrame(data_bat, columns=headers_bat)
batting_data
```

```python
batting_data.isna().sum()
```

```python
url_bowl = r'https://www.iplt20.com/stats/2021/mostWkts?stats_type=bowling'
html_bowl = urlopen(url_bowl)
soup_bowl = bs(html_bowl)

```

```python
headers_bowl = []
for head in soup_bowl.find_all('th')[2:-5] :
    headers_bowl.append(head.text)
headers_bowl
```

```python

data_bowl = []

for c in range(len(soup_bowl.find_all('tr'))) :
    if c<3 or c>=len(soup_bowl.find_all('tr'))-5: continue
    row = soup_bowl.find_all('tr')[c].find_all('td')
    l = []
    for x in row :
        l.append(x.text.strip())
    data_bowl.append(l)
```

```python
headers_bowl
```

```python
bowling_data = pd.DataFrame(data_bowl, columns=headers_bowl)
bowling_data
```
